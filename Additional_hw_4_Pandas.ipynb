```python
import datetime

def convert_time(date_string):
  try:
    if len(date_string) == 19 or len(date_string) == 18:
      return datetime.datetime.strptime(date_string, "%m.%d.%Y %H:%M:%S")
    elif len(date_string) == 16:
      return datetime.datetime.strptime(date_string, "%m.%d.%Y %H:%M")
    else:
      return date_string
  except ValueError as e:
      print(f"Помилка обробки: {date_string}. Помилка: {e}")
      return date_string

df['Timestamp'] = df['Timestamp'].apply(convert_time)
```

```python
top_10_in_object_columns = dict()

for i in df.select_dtypes('object').columns:
  value_counts = df[i].value_counts().sort_values(ascending=False).head(10)
  top_10_in_object_columns[i] = {value : count for value, count in value_counts.items()}

  print(f'Назва колонки: {i}')
  for top_value, top_count in top_10_in_object_columns[i].items():
    print(f'Значення: {top_value} - Частота: {top_count}')
```

13. Виведіть результат методу describe() по колонцы розміру ЗП для топ 5 за частотою зустрічання в даних областей (В якій області ви зараз живете?). Зробіть висновки, де вигідніше жити з точки зору ЗП і чи суттєва різниця на Ваш погляд для прийняття рішення про переїзд?

```python
# переглянемо частоту по всім областям (для контролю топ-5)
df['В якій області ви зараз живете?'].value_counts()
```

```python
#відбираємо топ-5 областей
top_5 = df['В якій області ви зараз живете?'].value_counts().head(5).index

#вібираємо частину дата-фрейму, який відноситься до обраних областей
filter = df[df['В якій області ви зараз живете?'].isin(top_5)]

#по відібраним даним будуємо групування та будуємо таблицю з описом (сортування зробимо по частоті області в порядку за спаданням значень).
# Також застосуэмо градієнт з попереднтого завдання, але з параметром axis=0
filter.groupby('В якій області ви зараз живете?')['Зарплата у $$$ за місяць, лише ставка після сплати податків'].describe().sort_values(by='count', ascending=False).style.background_gradient(axis=0, cmap='coolwarm').format('{:.0f}')
```

Результати обробки дата-фрейму показують, що нема великої доцільності переїзду наприклад, з Вінницької області до Києвської - рівень ЗП відрізняється незначною мірою (якщо не брати до уваги  максимальні значення даних). Відібрані дані також показують значну перевагу в кількості (фактично конкуренцію) опитаних спеціалістів в Києвській обалсті   в порівнянні з іншими облястями. Що також впливає на рішення переїзду до цієї області. Було б цікаво подивитися на кількість релевантних оголошень роботи по областях. Вважаю, що було б доцільним орієнтуватися ще на ці дані.
