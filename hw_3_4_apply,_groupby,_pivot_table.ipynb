```python
#Крок 2.

#Питаня по п.1 з Кроку 1

#Який розподіл по "Product line" та "Gender" покупців по кількості покупок
product_line_gender = pd.pivot_table(df, index=['Product line'], columns=['Gender'], values=['Invoice ID'], aggfunc= 'count')
product_line_gender.loc['Total count'] = product_line_gender.sum(numeric_only=True).values
product_line_gender = product_line_gender.rename(columns= {'Invoice ID' : 'Count'})
product_line_gender
```

```python
#Який розподіл по "Product line" та "Gender" покупців за показником валового доходу з їх продажів
product_line_gender_income = pd.pivot_table(df, index=['Product line'], columns=['Gender'], values=['gross income'], aggfunc=['sum']).round(2)
product_line_gender_income.columns = product_line_gender_income.columns.droplevel([0,1])

product_line_gender_income
```

```python
product_line_male = product_line_gender_income['Male'].sort_values(ascending=False).head(3)
product_line_male
```

```python
product_line_female = product_line_gender_income['Female'].sort_values(ascending=False).head(3)
product_line_female
```

```python
# Питаня по п.2 з Кроку 1

#Який розподіл по "Product line" та "Customer type" покупців по кількості покупок та за показником валового доходу в розрізі Branch
product_line_customer_type = pd.pivot_table(df, index=['Product line'], columns=[ 'Branch','Customer type'], values=['Invoice ID', 'gross income'], aggfunc={'Invoice ID' : 'count', 'gross income' : 'sum'}).round(2).astype(int)
product_line_customer_type.loc['Total'] = product_line_customer_type.sum(numeric_only=True).values
product_line_customer_type = product_line_customer_type.rename(columns={'Invoice ID': 'Number of Purchases', 'gross income': 'Total Income'})

product_line_customer_type
```

```python
# Питаня по п.3 з Кроку 1

#Який розподіл середнього значення "Rating" по "Gender" та "Customer type" в різних містах?
reiting_customer_type = pd.pivot_table(df, index= ['City'], columns= ['Customer type', 'Gender'], values=['Rating'], aggfunc= "mean", fill_value=0)
reiting_customer_type.loc['Total'] = reiting_customer_type.mean(numeric_only=True).values

reiting_customer_type
```
